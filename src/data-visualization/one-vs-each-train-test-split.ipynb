```python
import matplotlib.pyplot as plt 
import seaborn 


# pie with 1000 human texts, pie with 700 LLM_1 texts and 300 LLM_2 texts

```

```python
# declaring data 
data = [1, 1] 
keys = ['700 Human texts', '700 LLM_X texts'] 
  
# define Seaborn color palette to use 
palette_color = seaborn.color_palette(['blue', 'red']) 
  
# plotting data on chart 
plt.pie(data, labels=keys, colors=palette_color, startangle=90) 
  
# displaying chart 
plt.show() 
```
